```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numbers = [12, 15, 13, 14, 15, 15, 16, 17, 16, 18, 19, 18, 20, 22, 21, 23, 24, 25]

plt.figure(figsize=(6, 5))
sns.histplot(numbers, bins=6, kde=True)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = [200, 240, 300, 280, 310]

plt.figure(figsize=(8, 5))
sns.lineplot(x=days, y=sales, marker='o')
plt.title('Sales Over Days')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [50, 70, 30, 90]

plt.figure(figsize=(8, 5))
sns.barplot(x=categories, y=values)
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [3, 4, 2, 5, 6, 7]

plt.figure(figsize=(8, 5))
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot Example')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
```

```python
data = [12, 15, 14, 16, 13, 17, 14, 18, 19, 20]

plt.figure(figsize=(6, 4))
sns.boxplot(y=data)
plt.title('Seaborn Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B']
values = [12, 15, 14, 16, 13, 17, 14, 18, 19, 20]

plt.figure(figsize=(6, 4))
sns.boxplot(x=categories, y=values)
plt.title('Grouped Box Plot with Seaborn')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
data = [12, 15, 14, 16, 13, 17, 14, 18, 19, 20]

plt.figure(figsize=(6, 4))
plt.boxplot(data, vert=False, patch_artist=True)
plt.title('Matplotlib Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
df = pd.DataFrame({
    'Math': [80, 85, 90, 75, 95],
    'Science': [78, 83, 89, 70, 88],
    'English': [72, 76, 84, 65, 80]
})

corr = df.corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=4.0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df = pd.DataFrame({
'Height': [150, 160, 170, 180, 190],
'Weight': [50, 60, 70, 80, 90],
'Age': [20, 25, 30, 35, 40],
'Gender': ['F', 'M', 'F', 'M', 'M']
})

sns.pairplot(df, hue='Gender', diag_kind='hist')
plt.show()
```

```python
df = pd.read_csv('pokemon.csv')
```

```python
plt.figure(figsize=(24, 18))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Heatmap of Pokmon Stats Correlation")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='type1', y='attack', data=df)
plt.xticks(rotation=45)
plt.title("Attack Distribution by Pokmon Type")
plt.show()
```

```python
sns.pairplot(df[['hp', 'attack', 'defense', 'speed']], diag_kind='kde')
plt.suptitle("Pairplot of Key Pokmon Stats", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(x='is_legendary', y='attack', data=df)
plt.title("Attack Values by Legendary Status")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['speed'], fill=False)
plt.title("KDE Plot of Pokmon Speed")
plt.show()
```
