```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(70, 10, 100)  #70-mean    10-standard deviation   size=100
print(data)

plt.hist(data, bins=12, color="cyan", edgecolor='black')     #bin is number of stacs required to store X number of values/ Thichkness of bar
plt.title('Histogram of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 12]

plt.bar(categories, values, color='maroon')
plt.title('Sales by Product Category')
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.show()
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri']
temperatures = [30, 32, 33, 29, 35]

plt.plot(days, temperatures, marker='o', color='green')
plt.title('Temperature Over Days')
plt.xlabel('Days')
plt.ylabel('Temperature(*C)')
plt.grid(True)    #- shows or turns off the grid
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('pokemon.csv')

plt.figure(figsize=(8,5))
plt.hist(df['speed'], bins=10, color='pink', edgecolor='black')
plt.title('Distribution of Pokemon Speed')
plt.xlabel('Speed')
plt.ylabel('Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
avg_attack_by_gen = df.groupby('generation')['attack'].mean()

plt.figure(figsize=(8, 5))
plt.bar(avg_attack_by_gen.index.astype(str), avg_attack_by_gen.values, color='orange')

plt.title('Average Attack by Generation')
plt.xlabel('Generation')
plt.ylabel('Average Attack')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Get count of each primary type
type_counts = df['type1'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Pokémon Primary Types')
plt.legend(type_counts.index, title="Type 1", bbox_to_anchor=(1, 0.75))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sc = plt.scatter(df['attack'], df['defense'], c=df['generation'], cmap='Dark2', alpha=0.62, edgecolor='black')
#cmap - is the colormap
#alpha = sets transparency

plt.colorbar(sc, label='Generation')
plt.title('Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.grid(True)
plt.tight_layout()
plt.show()
```
