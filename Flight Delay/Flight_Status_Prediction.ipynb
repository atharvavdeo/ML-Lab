```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y = df['DepDel15']
```

```python
df = df.sample(20000, random_state=42)
```

```python
print(df.shape)  # rows, columns
print(df.memory_usage(deep=True).sum() / 1024**2, "MB")
```

```python
df = df.dropna(subset=['DepDel15'])

# Define target and features
y = df['DepDel15']
X = df.drop(columns=['DepDel15'])

# Encode categorical variables
cat_cols = X.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    X[col] = le.fit_transform(X[col].astype(str))

# Fill remaining NaNs
X = X.fillna(0)

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Train model with class balancing
log_model = LogisticRegression(max_iter=500, solver='saga', class_weight='balanced')
log_model.fit(X_train, y_train)

# Predict probabilities
y_probs = log_model.predict_proba(X_test)[:, 1]

import numpy as np
from sklearn.metrics import precision_recall_curve, confusion_matrix, classification_report

# Compute precision-recall curve
precisions, recalls, thresholds = precision_recall_curve(y_test, y_probs)

# Find threshold that balances precision and recall
best_threshold = thresholds[np.abs(precisions - recalls).argmin()]
print("Best Threshold:", best_threshold)

# Apply optimal threshold
y_pred_opt = (y_probs > best_threshold).astype(int)

# Evaluate
print("Confusion Matrix (Optimal Threshold):")
print(confusion_matrix(y_test, y_pred_opt))

print("\nClassification Report (Optimal Threshold):")
print(classification_report(y_test, y_pred_opt))
```

The Logistic Regression model achieved an overall accuracy of 73% after threshold tuning, with balanced performance between predicting on-time and delayed flights. However, the model struggled with delays specifically, achieving only 30% precision and 30% recall for the delayed class. This indicates that while the model can reasonably predict on-time flights, it is less effective at identifying delays—likely due to class imbalance, lack of feature scaling, and limited feature engineering.

Future improvements could include:
	•	Feature scaling for better convergence.
	•	Handling class imbalance (e.g., oversampling delays or using SMOTE).
	•	Using more advanced models (Random Forest or Gradient Boosting) that typically perform better on structured datasets like flight data.
